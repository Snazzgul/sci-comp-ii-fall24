```python
import matplotlib.pyplot as plt
import numpy as np

def absolute_value(val):
    a  = np.round(val/100.*y.sum(), 0)
    return a

y = np.array([36, 15, 16])
type = np.array(["Classes", "Research (unpaid)", "Teaching (paid)"])
myexplode = [0, 0, 0.1]
colors=np.array(['blue','teal','green'])
plt.pie(y, labels=type, colors=colors, autopct=absolute_value, explode = myexplode, shadow=True)
plt.title("Your First Two Years of Grad School, Hours per Week")
plt.show()
```

```python
y = np.array([36, 15, 16])
type = np.array(["12 Research Credits (unpaid)", "Addt'l. Research (unpaid)", "TA/RA Research (paid)"])
myexplode = [0, 0, 0.1]
colors= np.array(['teal','teal','green'])
plt.pie(y, labels=type, colors=colors,autopct=absolute_value,explode = myexplode, shadow=True)
plt.title("Your Last Four Years: Hours per Week")
plt.show()
```

```python
x = np.array([1,2,3,4,5,6])
w = np.array([16,16,52,52,52,52])
z = np.array([16,16,16,16,16,16])
plt.bar(x,w,alpha=0.4,color='teal',label='Hours OSU benefits from')
plt.bar(x,z,alpha=0.5,color='green', label='Hours we get paid for')
plt.title("How Much We Work vs. How Much We're Paid for")
plt.ylabel('Weekly Hours')
plt.xlabel('Year of Grad School')
plt.legend()
plt.show()
```
